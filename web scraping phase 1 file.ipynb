```python

breakfast1to=[]
breakfast2to=[]
breakfast3to=[]
breakfast_cusine1to=[]
Hotel_nametto=[]
No_of_reviewsto=[]
User_ratingsto=[]
Review_scoreto=[]
Locationto=[]



for i in page_url10:
    driver.get(i)
    time.sleep(2)
      
    namepa=driver.find_elements(By.XPATH,'//div[@class="hp__hotel-title pp-header"]/div/div/h2')
    for i in namepa:
        Hotel_nametto.append(i.text)
    
    breakfasts1=driver.find_elements(By.XPATH,'//div[@class="property_hightlights_wrapper"]/div[2]/div[3]/div[2]/p/span[1]')
    
    if breakfasts1:
        for i in breakfasts1: 
            breakfast1to.append(i.text.split(',')[0])
    else:
        breakfast1to.append('-')
           
    
    
    
    breakfasts2=driver.find_elements(By.XPATH,'//div[@data-testid="PropertyRestaurantsBlock-wrapper"]/section/div/div[2]/div/ul/li[2]/div/ul/li[1]/div/div[2]')
    if breakfasts2:
        for i in breakfasts2:
            
            breakfast2to.append(i.text.split(',')[0])
    else:
        breakfast2to.append('-')
    
    breakfasts3=driver.find_elements(By.XPATH,'//div[@data-testid="PropertyRestaurantsBlock-wrapper"]/section/div/div[2]/div/ul/li[3]/div/ul/li[1]/div/div[2]')
    if breakfasts3:
        for i in breakfasts3:
            
            breakfast3to.append(i.text.split(',')[0])
    else:
        breakfast3to.append('-')

            
    review=driver.find_elements(By.XPATH,'//div[@data-capla-component-boundary="b-property-web-property-page/PropertyReviewScoreRight"]/div/div/div[2]/div[2]')
    if review:
        for i in review:
            Review=i.text.replace('reviews','')

            No_of_reviewsto.append(Review)
    else:

        No_of_reviewsto.append('-')
    
    

    
    location= driver.find_elements(By.XPATH,'//p[@class="address address_clean"]/span[1]')
    for i in location:
        Locationto.append(i.text.split(',')[0])
                                                  
        
    User_rating=driver.find_elements(By.XPATH,'//div[@data-capla-component-boundary="b-property-web-property-page/PropertyReviewScoreRight"]/div/div/div[2]/div[1]')
    if User_rating:
        for i in User_rating:
            rates=i.text.replace('Review score','NA')
            User_ratingsto.append(rates)
    else :
        User_ratingsto.append('-')
            
                                                           
    score=driver.find_elements(By.XPATH,'//div[@data-capla-component-boundary="b-property-web-property-page/PropertyReviewScoreRight"]/div/div/div[1]')
    if score:    
        for i in score:
            
            Review_scoreto.append(i.text)
    else:
        Review_scoreto.append('-')
time.sleep(2)                                     
```
