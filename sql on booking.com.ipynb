# 1.   Write a query to give  details of hotels city is Sydney and  ratings is more than 5.

```python
hotel=cursor.execute("SELECT * FROM hotel_data WHERE Cities='Sydney' AND User_ratings> 5")
for i in hotel:
    print(i)
```

# 2.Write a query  to find the location with the most reviews

```python
city=cursor.execute('SELECT Cities FROM hotel_data GROUP BY Cities ORDER BY COUNT(*) DESC LIMIT 1')
for i in city:
    print(i)
```

# 3.Write a  query that  will retrieve all rows and sort them in descending order of user ratings.

```python
rating=cursor.execute('SELECT * FROM hotel_data ORDER BY User_ratings DESC')
for row in rating:
    print(row)
```

# 4.Write a query that  retrieves rows where the star rating is greater than or equal to 4.0 and sorts them in descending order of review score.



```python
rating2=cursor.execute('SELECT * FROM hotel_data WHERE Star_ratings>=4 ORDER BY User_ratings DESC')
for row in rating2:
    print(row)
```

# 5. Write a query to count the number of hotels in each city.

```python
citiess=cursor.execute("SELECT cities AS 'city', COUNT(*) 'No.of hotels' FROM hotel_data  GROUP BY cities ")
for i in citiess:
    print(i)
```
